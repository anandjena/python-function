9. Explain the purpose and usage of the `map()` function in Python.
- The map() function in Python is a built-in function that applies a given function to all items in an iterable such as a list, tuple, or string and returns an iterator that produces the results. The purpose of map() is to transform each item in an iterable by applying a function to each of them, without having to explicitly loop through the iterable yourself.

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- The map() function is used to apply a given function to every item in an iterable (like a list or tuple) and returns a new iterator that yields the transformed items.

  The reduce() function, available in the functools module, reduces an iterable to a single cumulative value by applying a function that takes two arguments at a time and accumulates the result.

  The filter() function is used to filter out items from an iterable based on a condition defined by a function that returns True or False. It returns an iterator of items for which the function evaluates to True.

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
my_list = [1,2,4,5,7,8,10,14,23,21,45]

def my_func(l):
    sum_in = 0
    for i in l:
        if i%2 == 0:
            sum_in += i
    return sum_in

my_func(my_list)
```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string.
def my_func(s):
    if type(s) == str:
        return s[::-1]

my_func("hello")
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
my_list = [1,2,4,5,7,8,10,14,23,21,45]

def my_func(list_val):
  return list(map(lambda x:x**2,list_val))

my_func(my_list)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def my_func(n):
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
my_func(100)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fib(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a+b

for i in fib(10):
    print(i)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def fib(n):
  for i in range(n):
    yield 2**i

for i in fib(10):
  print(i)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

for line in read_file('sample.txt'):
    print(line)
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

my_list = [(1, 2), (4, 1), (9, 10), (13, -3)]

sorted_list = sorted(my_list , key = lambda x : x[1])
print(sorted_list)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temps = [0, 20, 30, 37, 100]

list(map(lambda x : (x * 9/5) + 32,celsius_temps))
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
vowels = 'aeiouAEIOU'
def funcc(stri):
    return ''.join(filter(lambda x : x not in vowels,stri))
funcc("my name is anand")
```
